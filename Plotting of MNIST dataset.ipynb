```python
num = 10
images = X_train[:num]
labels = Y_train[:num]

num_row=2
num_col=5

fig,axes = plt.subplots(num_row,num_col,figsize = (1.5*num_col,1.5*num_row))

for i in range(num):
    ax = axes[i//num_col,i%num_col]
    ax.imshow(images[i],cmap='gray')
    ax.set_title('Label: {}'.format(labels[i]))
    
plt.tight_layout()
plt.show()
```
